```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1

```python
complete=pd.read_csv(r"C:\Users\\Hp\\Downloads\covid_19_clean_complete.csv")
```

```python
complete
```

```python
complete.dtypes
```

```python
complete.replace('NaN',0,inplace=True)
complete
```

```python
complete['Province/State']=complete['Province/State'].fillna('Unknown').astype(object)
complete['Date']=pd.to_datetime(complete['Date'],errors='coerce')
complete
```

```python
from matplotlib.pyplot import figure
figure(num=None, figsize=(15,10), dpi=40, facecolor='w', edgecolor='r')
sns.barplot(x='Country/Region', y = 'Active' ,data = complete)
plt.show()
```

```python
h=complete
x=h['WHO Region']
y=h['Confirmed']
f=h['Recovered']
plt.bar(x,y,label='Confirmed')
plt.bar(x,f,label='Recovered')
plt.legend()

plt.show()
```
